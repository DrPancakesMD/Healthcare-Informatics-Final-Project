```python
import pandas as pd
import numpy as np
```

Creating function to change Y/N to boolean

```python
def convert_bool(col):
    if str(col).title() ==  "Y": #check for nan
        return True
    elif str(col).title() == "N":
        return False
    else:
        return col
```
