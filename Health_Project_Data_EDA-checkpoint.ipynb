```python
pca_1 = pc[['percent_poverty_state','Measure Performance Rate', 'Measure Title']]

pca_1.replace([np.inf, -np.inf], np.nan)
pca_1.dropna(inplace= True)
pca_1.isnull().values.any()
pca_1.isnull().sum()

#pca_1
```

```python
#https://towardsdatascience.com/pca-using-python-scikit-learn-e653f8989e60

from sklearn.preprocessing import StandardScaler
features = ['percent_poverty_state', 'Measure Performance Rate']
# Separating out the features
x = pca_1.loc[:, features].values
# Separating out the target
y = pca_1.loc[:,['Measure Title']].values
# Standardizing the features
x = StandardScaler().fit_transform(x)
x

from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalComponents
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])

#principalDf

y = pd.DataFrame(data = y, columns = ['Measure Title'])
#y

finalDf = pd.concat([principalDf, y], axis = 1)


fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = ['Prevent Flu', 'Prevent Tobacco', 'Osteo Screen Female', 'Screen HCV']
colors = ['red', 'blue', 'green', 'black']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['Measure Title'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

# Linear Regression

## Flu Prevention - Percent Poverty by State

```python
pc.dropna(inplace= True)

flu_rows = pc[pc['Measure Title'] == 'Prevent Flu'].dropna(subset=['Measure Performance Rate', 'percent_poverty_state'])
flu_rows.drop_duplicates(subset=['NPI', 'Measure Performance Rate', 'zip'], keep='first', inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(flu_rows['Measure Performance Rate'].values.reshape(-1, 1),
                                                    flu_rows['median_household_zip'].values.reshape(-1, 1),
                                                    test_size = .25, random_state=25)
```

```python
LinReg = LinearRegression()
LinReg.fit(X_train, y_train)
y_pred = LinReg.predict(X_test)

# The coefficients
print('Coefficients: \n', LinReg.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, y_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y_test, y_pred))

# Plot outputs
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```
